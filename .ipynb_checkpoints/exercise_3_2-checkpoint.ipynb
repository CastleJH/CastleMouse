In this part of the lecture we explain Stochastic Gradient Descent (SGD) which is an **optimization** method commonly used in neural networks. We will illustrate the concepts with concrete examples.

시작하기 전에 SGD가 뭔지 찾아봤다. 우선 Gradient Descent는 경사하강법이라는 것으로, 손실을 줄이는 알고리즘이다. 가중치(w)에 따른 손실(l)이 어떻게 변화하는지 함수를 그리면 우리는 그 함수의 기울기가 최소가 되는 점을 찾아 손실을 줄일 수 있다(보통 이 함수다image.png). 아래는 경사 하강법의 단계이다.

1. w1의 시작점을 선택한다. 보통 0으로 설정하는데 시작점이 크게 중요하진 않다고 한다. 근데 만약 cost 함수가 위 사진과 같지 않으면 시작값을 찾는것도 매우 중요하다. 일단 cost함수가 위와 같다고 하겠다.

2. 손실 곡선의 시작점에서의 기울기를 계산한다. **여기서 learning rate가 나온다.** 기울기를 계산할 때 시작점의 다음 지점을 결정하는게 learning rate(기울기의 보폭)이다. (아직도 learning rate가 뭔지 감이 잘 안왔는데 드디어 알게 되었다.)

3. 위 과정을 반복하면서 최소값에 점점 접근한다.

그렇다면 SGD(확률적 경사하강법)는 뭘까? 이것은 배치 크기가 1인 경사하강법 알고리즘이다. 배치란 단일 반복에서 기울기를 게산하는데 사용되는 data의 총 개수이다. 만약 dataset이 엄청나게 크다면 배치가 엄청 거대해질 것이고 결국 단일반복으로도 계산하는데 오랜 시간이 걸릴 수 있다. 그리고 대량의 데이터셋에서는 중복 데이터가 포함되어 있을 가능성이 커진다. 그래서 데이터셋에서 우리가 data를 무작위로 선택해주면 훨씬 적은 데이터셋으로 우리가 원하는 값을 찾을 수 있다. 바로 이것이 SGD가 배치 크기가 1이라는 것(반복당 하나의 예만을 사용)의 의미이다.

물론 단점이 존재한다. 일단 최저점에 가까운 점을 찾을 가능성이 높긴 해도, 항상 보장되는 것은 아니다(dataset에서 무작위를 뽑아서 하기 때문에). 그래서 이 단점을 극복하기 위한 절충안이 미니 배치 SGD이다. 이거는 배치가 1이 아니라 10~1000개 사이의 예로 구성된다. SGD의 단점을 보충하면서도 전체 배치보다는 효율적이다.

#  Linear Regression problem

The goal of linear regression is to fit a line to a set of points.

우선 정말 간단한 선형 model을 만들 것이다. x축은 온도이고, y축은 우리가 파는 아이스크림 개수라고 생각하면 된다. 물론 실제로 아이스크림과 온도의 관계는 아니고, 비슷한 여러가지 경우가 있으니 원하는 걸로 여기면 된다. 

```python
n=100
```

100개의 데이터가 있다.

```python
x = torch.ones(n,2) 
x[:,0].uniform_(-1.,1)
x[:5]
```

```python
a = tensor(3.,2); a
```

우리가 고른 계수는 [3, 2]이다. 나중에 이걸 예측해볼 것이다.

```python
y = x@a + torch.rand(n)
```

위에서 정한 a를 x와 행렬 곱셈 해주고, 랜덤한 값을 더해주면서 데이터셋을 만들었다.

```python
plt.scatter(x[:,0], y);
```

You want to find **parameters** (weights) `a` such that you minimize the *error* between the points and the line `x@a`. Note that here `a` is unknown. For a regression problem the most common *error function* or *loss function* is the **mean squared error**. 

```python
def mse(y_hat, y): return ((y_hat-y)**2).mean()
```

위 코드는 이 수식이다.

image.png

Suppose we believe `a = (-1.0,1.0)` then we can compute `y_hat` which is our *prediction* and then compute our error.

```python
a = tensor(-1.,1)
```

우리는 이제 a를 예측해보고 싶다. 물론 우리는 앞에서 [3, 2]라고 직접 줬기 때문에 정답을 알지만, 모른다고 가정하고 일단 -1, 1로 계산해본다.

```python
y_hat = x@a
mse(y_hat, y)
```

```python
plt.scatter(x[:,0],y)
plt.scatter(x[:,0],y_hat);
```

파란색은 실제 데이터 분포고, 주황색은 a가 [-1, 1]이라고 하였을 때의 개형이다. 매우 틀렸다. 그럼 이제 파란선으로 가까워지도록 주황선을 수정해줘야 하는데, 이 과정을 수행할 update라는 함수를 밑에서 만들었다. 이 update 함수를 계속 적용시켜주면서 주황색을 파란선에 가장 가깝게 만들어주는 a 값을 찾도록 하겠다.

So far we have specified the *model* (linear regression) and the *evaluation criteria* (or *loss function*). Now we need to handle *optimization*; that is, how do we find the best values for `a`? How do we find the best *fitting* linear regression.

# Gradient Descent

We would like to find the values of `a` that minimize `mse_loss`.

**Gradient descent** is an algorithm that minimizes functions. Given a function defined by a set of parameters, gradient descent starts with an initial set of parameter values and iteratively moves toward a set of parameter values that minimize the function. This iterative minimization is achieved by taking steps in the negative direction of the function gradient.

Here is gradient descent implemented in [PyTorch](http://pytorch.org/).

```python
a = nn.Parameter(a); a
```

```python
def update():
    y_hat = x@a
    loss = mse(y, y_hat)
    if t % 10 == 0: print(loss)
    loss.backward()
    with torch.no_grad():
        a.sub_(lr * a.grad)
        a.grad.zero_()
```

```python
lr = 1e-1
for t in range(100000): update()
```

update를 100번 반복해서 a값을 수정했다.

```python
print(a)
plt.scatter(x[:,0],y)
with torch.no_grad():
    plt.scatter(x[:,0],x@a);
```

그 결과 제대로 작동했다. 
a도 얼추 비슷하게 예측하였다. 강의자분은 100번 돌렸는데 나는 200번, 1000번, 그리고 마지막으로 십만번을 돌렸다. 조금씩 a값이 [3, 2]에 더 가까워지긴 했지만 갈수록 수정되는 효율이 줄어들었다. 이래서 epoch를 너무 크게 하면 안되는 것이다. 시간만 많이 들고 효과는 높지 않으니까!

## Animate it!

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
a = nn.Parameter(tensor(-1.,1))

fig = plt.figure()
plt.scatter(x[:,0], y, c='orange')
with torch.no_grad():
    line, = plt.plot(x[:,0], x@a)
plt.close()

def animate(i):
    update()
    with torch.no_grad():
        line.set_ydata(x@a)
    return line,

animation.FuncAnimation(fig, animate, np.arange(0, 100), interval=20)
```
