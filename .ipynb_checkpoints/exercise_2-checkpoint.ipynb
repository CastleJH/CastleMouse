The first thing we do when we approach a problem is to take a look at the data. We _always_ need to understand very well what the problem is and what the data looks like before we can figure out how to solve it. Taking a look at the data means understanding how the data directories are structured, what the labels are and what some sample images look like.

The main difference between the handling of image classification datasets is the way labels are stored. In this particular dataset, labels are stored in the filenames themselves. We will need to extract them to be able to classify the images into the correct categories. Fortunately, the fastai library has a handy function made exactly for this, `ImageDataBunch.from_name_re` gets the labels from the filenames using a [regular expression](https://docs.python.org/3.6/library/re.html).

```python
fnames = get_image_files(path_img)
fnames[:5]
```

docs에서 get_image_files에 대해 다음과 같이 설명했다.

get_image_files(c:PathOrStr, check_ext:bool=True, recurse=False) → FilePathList

Return list of files in c that are images. check_ext will filter to image_extensions

그래서 위에서는 현재 get_image_files가 path_img로부터 이미지인 파일들의 리스트를 반환했다.

리스트에서 앞에 5개 정도 뭐가 있나 봤다. jpg파일이 들어있다.

```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

fnames에 들어있는 이미지 파일 이름에서 label을 뽑아낼 것이다. (label이 바로 우리가 예측하려는 것이다.)

우리는 pat의 정규표현식으로 파일 이름에서 label을 뽑아올거다.

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size=224, bs=64)
data.normalize(imagenet_stats)
```

fastai에는 ImageDataBunch라는 오브젝트가 있는데, 이건 label이랑 이미지가 붙은 training, validation, 그리고 경우에 따라 test 데이터 까지 모두 포함하고 있다.

이렇게 데이터셋 만드는 방법은 여러개가 있는데(from_folder, from_csv, from_df, from_name_func 등등... 맨 밑에 3과 7을 구분하기 위한 ImageDataBunch를 만들 때 예시를 보여줄 것이다.) 위에서 말했듯이 우리는 정규표현식으로 label을 붙일 것이다. 그래서 from_name_re를 썼다.

docs에선 from_name_re를 다음과 같이 설명했다.

from_name_re(path:PathOrStr, fnames:FilePathList, pat:str, valid_pct:float=0.2, **kwargs)

Create from list of fnames in path with re expression pat.

이에 따라, 위 코드로 우리는 변형이 실행된(get_transforms()를 default로 뒀기 때문에 지금 일부 사진들은 좌우반전, 회전, 뒤틀림, 색조변경이 적용되었다) 사이즈 224짜리 이미지로 이루어진 데이터셋이 만들어졌다. 이미지 사이즈나 비율이 다 다르면 GPU로 같은 작업을 굉장히 빠르게 반복하고 학습하기 힘들다. 그래서 224크기의 정사각형으로 모두 모양을 같게 고정시킨 것이다. 224인 이유는 마지막 레이어가 7x7사이즈기 때문이라고 한다. (7*2^5)

## Training: resnet34

Now we will start training our model. We will use a [convolutional neural network](http://cs231n.github.io/convolutional-networks/) backbone and a fully connected head with a single hidden layer as a classifier. Don't know what these things mean? Not to worry, we will dive deeper in the coming lessons. For the moment you need to know that we are building a model which will take images as input and will output the predicted probability for each of the categories (in this case, it will have 37 outputs).

We will train for 4 epochs (4 cycles through all our data).

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate) 
```

이제 우리는 model을 train할 수 있다. fastai에서는 Learner라는 클래스가 있는데 이걸 활용에서 train이 가능하다고 한다.

cnn은 convolutional neural network의 약자다. 이미지에서 객체, 얼굴, 장면등을 인식하기 위해 패턴을 찾는데에 특히 유용한 알고리즘이라고 한다. 

이제 cnn을 위한 learner를 만드는데에는 두가지가 필요하다. 바로 data랑 ResNet이다. 강의에서 이게 되게 좋은 모델이라고 하면서 도표로 비교해주기까지 했다. 어쨌든 ResNet에 몇 장의 layers가 있느냐에 따라 resnet뒤에 숫자가 붙는데, 시간을 절약하기 위해 여기서는 34를 쓸 것이다. 

https://docs.fast.ai/vision.learner.html#cnn_learner

cnn_learner(data:DataBunch, base_arch:Callable, cut:Union[int, Callable]=None, pretrained:bool=True, lin_ftrs:Optional[Collection[int]]=None, ps:Floats=0.5, custom_head:Optional[Module]=None, split_on:Union[Callable, Collection[ModuleList], NoneType]=None, bn_final:bool=False, init='kaiming_normal_', concat_pool:bool=True, **kwargs:Any) → Learner

일단, 위 코드로 learn에는 나의 data에 맞는 Learner 오브젝트가 있다.



> ## 저희는 fast.ai가 제공하는 모델을 그대로 가져와 사용합니다. 그리고 이 모델은 이미 resnet34를 가지고 어느정도 basic한 부분들을 학습을 해놓은 상태입니다.
> ## 그리고 그 모델을 저희가 classification하려는 data에 적용할 모델을 learn에 선언한 것이죠.

```python
learn.fit_one_cycle(4)
```

이제 학습을 시켜줄건데, 4번 시켜줄 것이다. 1번만 하고 끝내지 않는 이유는, 당연하게도, 여러번 학습시켜줄 수록 더 나아질 것이기 때문이다. 하지만 너무 많이 반복하면 시간도 오래 걸리고 overfitting이 발생할 수 있다. 그냥 그 특정 picture만 인지하고 pet을 전체적으로 구분할 수 없을 것이다. 그래서 4번을 돌렸다.

## Unfreezing, fine-tuning, and learning rates

Since our model is working as we expect it to, we will *unfreeze* our model and train some more.

```python
learn.unfreeze()
```

이제 model을 더 업그레이드 시킬 것이다. 그전에 우리가 fit_one_cycle을 불렀었다. 그리고 cnn에는 layer가 많은데, 이전에 우리가 돌린 건 사실 뒤에 레이어 몇장을 추가해서 이것들만 train한 것이었다. 

따라서 시간은 좀 더 걸려도 전체 model을 train하기 위해서 우리는 unfreeze를 명령한 후에 fit_one_cycle을 불러야 한다.

```python
learn.fit_one_cycle(1)
```

그런데 별로 효과가 없어보인다. unfreeze전에는 에러율이 첫 사이클에서 0.10이었는데 말이다.

그 이유는 다음과 같다. **(여기서부터는 내가 확실히 이해한것 같지 않다.)**

cnn은 여러 layer로 이루어져있다. 1번째 layer의 coefficients로 대각선, 황색에서 청색으로의 변화 등 매우 간단한 것들을 인식하게 된다. 2번째 layer는 1번째를 통해 좀더 specific한 것을 알게 된다. 예를들면 두개의 선이 합쳐진 직각이라든지. 1번째 layer에서 하나의 선을 구분하고 색깔 변화를 인식했다면, 2번째 layer에서는 그 결과를 활용해 두개의 선을 인식하고, 더많은 색변화를 인식하는 것이다. 그다음, 3번째 layer는 다시 이전 결과들로 더 specific한 것을, 4번째에서는 더, 5번째에서는 더더... 이런식으로 이전 레이어의 결과들을 결합해서 다음 레이어에서는 더 특정한 형태를 구분해낼 수 있게 된다.

우리가 처음 train 했을 때 마지막에 우리가 좀 더 추가해준 몇 layers를 제외하면, 이전 layers는 이미 ResNet34에서 구한 coefficients나 parameters가 있는 상태였다. 우리가 한건 그저 이미 값이 미리 구해진 ResNet34의 layers에 새로 몇개 layers를 더해서 그들의 coefficients및 parameters를 구한 것이다.

그런데 우리는 이번엔 중간의 모든 layers까지 다 뜯어서 우리가 구하고자 하는것에 더 부합하는 값을 찾고자 한다. 1번째 layer는 매우 단순한 형태니 딱히 고칠 필요가 없겠지만, 더 높은 layer로 갈 수록 다른 방향으로 고쳐질 것이다. 

바로 이것 때문에 위 코드로 딱히 좋은 결과를 갖지 못한 것이다. 왜냐하면 default값에 의해(0.003) 모든 layer들이 지금 구분하는게 얼마나 단순한 형태인지 복잡한 형태인지에 상관 없이 똑같은 스피드(같은 learning rate)로 고쳐진다. 그러니 layer들이 수정하는 learning rate를 서로 달리 해주면 의미있는 방향으로 model을 업뎃할 수 있다. 


```python
learn.load('stage-1')
```

이제 원인을 알았으니, 의미없게 수정된 가중치들을 다시 원래대로 돌려놓고 싶다. unfreeze해서 fitting하기 전의 값들을 우리는 이전에 저장해놓았다. 위 코드로 그 상태를 불러왔다.

```python
learn.lr_find()
learn.recorder.plot()
```

아까 learning rate를 달리하겠다고 했다. 그러면 우리는 이제 가장 효율적인 수준을 찾아야 한다. 위 코드들로 그래프를 통해 learning rate에 따른 loss가 어떻게 되는지 확인 할 수 있었고, 대략 1e-4보다 learning rate가 올라가자 loss가 빠르게 증가함을 알 수 있었다. docs에의하면 default가 0.003이었다. 이 지점은 loss가 한참 증가한 지점이었다.

fit_one_cycle(learn:Learner, cyc_len:int, max_lr:Union[float, Collection[float], slice]=slice**(None, 0.003, None)**, moms:Point=(0.95, 0.85), div_factor:float=25.0, pct_start:float=0.3, final_div:float=None, wd:float=None, callbacks:Optional[Collection[Callback]]=None, tot_epochs:int=None, start_epoch:int=None)

```python
learn.unfreeze()
learn.fit_one_cycle(2, max_lr=slice(1e-6,1e-4))
```

이제 가장 좋은 learning rate도 알았으니 다시 unfreeze하고 fitting해준다. 여기서 우리는 slice를 통해 가장 첫번째 layer는 1e-6의 rate로, 뒤로갈수록 점점 1e-4에 가까워지다 마지막에는 1e-4의 rate로 fitting하도록 명령한다. 

**Q. 여기서 가장 이해가 안된다. 뒷 layer로 갈 수록 learning rate가 느려야 앞 layer의 수정 결과가 뒤에 반영되는 것 아닌가? 왜 뒤가 더 높은 learning rate를 가지는 걸까? 내가 아무래도 앞부분에서 이해가 미진한 것 같다. - https://github.com/hiromis/notes/blob/master/Lesson1.md 의 1:22:58부분**


기다린 결과, 어느때보다도 좋은 수준의 에러율이 나왔다!

> ## learning rate는 한 번 학습할 때 얼마나 학습하는지를 의미합니다. 매우 간단한 구조의 앞 layer들은 미세한 부분이기 때문에, 더 많이 학습을 해주어야 디테일한 부분을 캐치할 수 있겠죠.
> ## 반대로, 뒤의 layer들은 좀 더 고차원적인 부분들이 많아서 더 많은 data들을 다루면 좋기 때문입니다.

That's a pretty accurate model!

## Training: resnet50

Now we will train in the same way as before but with one caveat: instead of using resnet34 as our backbone we will use resnet50 (resnet34 is a 34 layer residual network while resnet50 has 50 layers. It will be explained later in the course and you can learn the details in the [resnet paper](https://arxiv.org/pdf/1512.03385.pdf)).

Basically, resnet50 usually performs better because it is a deeper network with more parameters. Let's see if we can achieve a higher performance here. To help it along, let's us use larger images too, since that way the network can see more detail. We reduce the batch size a bit since otherwise this larger network will require more GPU memory.

```python
data = ImageDataBunch.from_name_re(path_img, fnames, pat, ds_tfms=get_transforms(), size= 299, bs=32)
data.normalize(imagenet_stats)
```

여기서부터는 앞과 같은데 그냥 ResNet50을 쓴다. 다만 내 GPU문제로 bs를 이전보다 낮췄다.

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

이전에는 34였지만 이번엔 ResNet50으로 Learner 오브젝트를 만들었다.

```python
learn.fit_one_cycle(4)
```

강의에서는 사이클을 5번 돌렸는데 34에 비해 너무 오래걸려서 그냥 4번만 돌리기로 했다.
역시, 시간이 훨씬 오래 걸리긴 해도 에러율은 5.3으로 34를 썼을 때의 6.9보다 낮다.

```python
learn.save('stage-1-50')
```

이번에도 unfreeze하고 모든 layers를 수정할 것이다.

It's astonishing that it's possible to recognize pet breeds so accurately! Let's see if full fine-tuning helps:

```python
learn.unfreeze()
learn.fit_one_cycle(1, max_lr=slice(1e-6,1e-4))
```

아까 우리는 어느정도의 learning rate가 적당한지 알았으니 이번에도 아까의 learning rate를 써서 fitting하겠다.

사이클을 1번 돌렸을 때의 결과가 아까보다 더 좋아졌다. 에러율이 5.6밖에 되지 않는다.

If it doesn't, you can always go back to your previous model.

```python
'''learn.load('stage-1-50');'''
```

결과가 별로 나아지지 않았다면 이부분을 실행해서 돌아가면 되긴 하다만, 굳이 돌아가지 않아도 될 것 같다.

```python
interp = ClassificationInterpretation.from_learner(learn)
```

이번에도 시각화해서 어느 케이스가 가장 예측이 힘들었는지 볼 것이다.

```python
interp.most_confused(min_val=2)
```

ResNet34를 쓸 때와 케이스가 어느정도 겹치는 것 같다. 그리고 위의 혼동 케이스에 따르면 같은 고양이 안에서, 또는 개 안에서는 어떤 종인지 혼동하는 경우가 있지만 고양이와 개를 구분하는 것은 거의 완벽하다!

```python
data = ImageDataBunch.from_csv(path, ds_tfms=tfms, size=28)
```

from_csv는 csv파일을 통해 DataBunch를 만든다. 이 경우에는 label이 3이나 7이 아닌, 0과 1로 이루어져있다. 이게 뜻하는 것은 "7인가 아닌가?"이다(7이면 true(1) 3이면 false(0)를 의미).


참고로 만약에 파일 이름이 'labels.csv'이면 굳이 다른 파일이름을 넘겨줄 필요가 없다. 이게 default값이기 때문이다.

from_csv(path:PathOrStr, folder:PathOrStr=None, label_delim:str=None, csv_labels:PathOrStr=**'labels.csv'**, valid_pct:float=0.2, seed:int=None, fn_col:int=0, label_col:int=1, suffix:str='', delimiter:str=None, header:Union[int, str, NoneType]='infer', **kwargs:Any) → ImageDataBunch

```python
data.show_batch(rows=3, figsize=(5,5))
data.classes
```

data에 무엇이 들어있는지 3x3으로 본다. 그리고 label이 0과 1로 이루어져있는 것도 확인했다.

```python
data = ImageDataBunch.from_df(path, df, ds_tfms=tfms, size=24)
data.classes
```

from_csv가 csv파일을 참조했다면, from_df는 DataFrame을 참조한다. 나머지는 from_csv와 같다.

```python
fn_paths = [path/name for name in df['name']]
fn_paths[:2]
```

```python
pat = r"/(\d)/\d+\.png$"
data = ImageDataBunch.from_name_re(path, fn_paths, pat=pat, ds_tfms=tfms, size=24)
data.classes
```

이전에 개와 고양이의 경우처럼 정규 표현식으로 label을 붙인다.

```python
data = ImageDataBunch.from_name_func(path, fn_paths, ds_tfms=tfms, size=24,
        label_func = lambda x: '3' if '/3/' in str(x) else '7')
data.classes
```

이번엔 from_name_func로 함수를 통해 label을 붙인다. 람다함수를 써서 경로에/3/이 있으면 label을 3으로 하고, 아니면 7로 한다.

```python
labels = [('3' if '/3/' in str(x) else '7') for x in fn_paths]
labels[:5]
```

labels이라는 리스트에다가 경로에 /3/이 있으면 3을, 아니면 7을 저장한다.

```python
data = ImageDataBunch.from_lists(path, fn_paths, labels=labels, ds_tfms=tfms, size=24)
data.classes
```

from_lists로 위의 labels리스트를 활용해서 label을 붙인다.
